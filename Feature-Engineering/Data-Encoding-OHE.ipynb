## Nominal/OHE Encoding

One-hot encoding also known as Nominal encoding, is a technique used to represent categorical data as numerical data, which is more suitable for ML algorithms.

In this technique, each category is represented as a binary vector where each bit corresponds to a unique category.

For example, If we have categorical variable `color` with 3 possible values (red, green, blue) then we can represent it in OHE encoding as below:
- Red: [1, 0, 0]
- Green: [0, 1, 0]
- Blue: [0, 0, 1]

```python
# Create a simple dataframe'
df = pd.DataFrame({
    'color': ['red', 'green', 'blue', 'green', 'red', 'blue']
})
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df.head()
```

```python
# Create an instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoded_df.head()
```

```python
# Transform new data in future with encoder
encoder.transform([['blue']]).toarray()
```
