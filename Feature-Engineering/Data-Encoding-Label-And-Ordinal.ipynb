# Label Encoding

Label encoding and ordinal encoding are two techniques used to encode categorical data as numerical data.

Label encoding involves assigning a unique numerical label to each category in the variable.
The labels are usually assigned in alphabetical order or based on frequency.

For example, if we have categorical variable `color` with 3 possible values (red, green, blue) then we can represent it using label encoding as below:

1. Red: 1
2. Green: 2
3. Blue: 3

```python
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'green', 'red']
})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lbl_encoder = LabelEncoder()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

Here we can see, (Red: 2), (Green: 1), and (Blue: 0)

```python
# We can get encoded value for data
lbl_encoder.transform([['red']])
```

```python
# We can't get encoded value for unseen data during `fit`
# It will throw an error
# lbl_encoder.transform([['yellow']])
```

# Ordinal Encoding

It is used to encode categorical data which has intrinsic order or ranking.

In this technique, each category is assigned a number based on its position in the order.

For example, if we have categorical variable `education_level` with 4 possible values (high school, college, graduate, post graduate) then we can represent it as below:

1. High School:  1
2. College: 2
3. Graduate: 3
4. Post Graduate: 4

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df2 = pd.DataFrame({
    'size': ['small','medium', 'large', 'medium', 'small', 'large']
})
```

```python
ord_encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
ord_encoder.fit_transform(df2[['size']])
```

```python
ord_encoder.transform([['small']])
```
