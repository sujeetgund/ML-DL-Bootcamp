### Features  
Here are the key features of the cleaned dataset:  

1. **Airline**: Identifies the airline company, with 6 unique airlines included.  
2. **Flight**: Represents the flight code, stored as a categorical feature.  
3. **Source City**: Specifies the city where the flight departs, with 6 unique cities.  
4. **Departure Time**: Groups departure times into 6 time slots, providing a clear classification of flight schedules.  
5. **Stops**: Indicates the number of stops between the source and destination, with 3 distinct categories.  
6. **Arrival Time**: Groups arrival times into 6 time slots, helping to categorize flight arrival schedules.  
7. **Destination City**: Specifies the city where the flight lands, with 6 unique cities represented.  
8. **Class**: Indicates the seat class, categorized as either Business or Economy.  
9. **Duration**: Shows the total travel time (in hours) between the source and destination.  
10. **Days Left**: Calculates the number of days between the booking date and the travel date, providing insight into booking timelines.  
11. **Price**: Represents the target variable, capturing the flight ket price.

```python
# Decode `Date_of_Journey` column
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] =  df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
# Now we can drop `Date_of_Journey` column
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.sample(5)
```

```python
# Get rid of date data from `Arrival_Time` column
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df.head(5)
```

```python
df['Arrival_Hours'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Mins'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.sample(5)
```

```python
df['Arrival_Hours'] = df['Arrival_Hours'].astype(int)
df['Arrival_Mins'] = df['Arrival_Mins'].astype(int)
```

```python
# Drop `Arrival_Time` column
df.drop('Arrival_Time', axis=1, inplace=True)
```
